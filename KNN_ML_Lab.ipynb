# KNN Classification using new_dataset.csv
This notebook loads the same dataset, encodes categorical columns, trains a KNN model, and checks accuracy.

```
import numpy as np
import pandas as pd
```

```
# Load dataset (make sure file is uploaded in Colab)
df = pd.read_csv("new_dataset.csv")
df.head()
```

```
df.isnull().sum()
```

```
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

# Encode categorical columns (same as your DT experiment)
for col in ['Temp','Humidity','Windy','Outlook']:
    if col in df.columns:
        df[col] = lb.fit_transform(df[col])

df.head()
```

```
X = df.drop(columns=['Play'], axis=1)
y = df['Play']
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=30
)
```

```
from sklearn.neighbors import KNeighborsClassifier

# Create model (k=3 usually good for small datasets)
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(x_train, y_train)
```

```
from sklearn.metrics import accuracy_score

y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred)
```
