# Naive Bayes Classification using new_dataset.csv
Load dataset, encode categorical values, train Naive Bayes model, and compute accuracy.

```
import numpy as np
import pandas as pd
```

```
# Upload new_dataset.csv in Colab first
df = pd.read_csv("new_dataset.csv")
df.head()
```

```
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

# encode same columns as previous experiments
for col in ['Temp','Humidity','Windy','Outlook','Play']:
    if col in df.columns:
        df[col] = lb.fit_transform(df[col])

df.head()
```

```
X = df.drop(columns=['Play'], axis=1)
y = df['Play']
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=30
)
```

```
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```
from sklearn.metrics import accuracy_score

y_pred = nb_model.predict(x_test)
accuracy_score(y_test, y_pred)
```
